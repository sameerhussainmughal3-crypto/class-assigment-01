Our friend has also been brushing up on their Python skills and has taken a first crack at a CSV file containing Netflix data. For their first order of business, they have been performing some analyses, and they believe that the average duration of movies has been declining. </p>
<p>As evidence of this, they have provided us with the following information. For the years from 2011 to 2020, the average movie durations are 103, 101, 99, 100, 100, 95, 95, 96, 93, and 90, respectively.</p>
<p>If we're going to be working with this data, we know a good place to start would be to probably start working with <code>pandas</code>. But first we'll need to create a DataFrame from scratch. Let's start by creating a Python object covered in <a href="https://learn.datacamp.com/courses/intermediate-python">Intermediate Python</a>: a dictionary!</p>


```python
sample_year = [2011 , 2012 , 2013 , 2014 , 2015 , 2016 , 2017 , 2018 , 2019 , 2020]
sample_duration = [103, 101 , 99, 100, 100 ,95, 95, 96, 93, 90]
sample_movie_dict = {
   'year':sample_year,
   'duration':sample_duration
}
sample_movie_dict
```

```python
import pandas as pd
```

```python
sample_duration_df = pd.DataFrame(sample_movie_dict)
sample_duration_df
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12, 5))
plt.plot(sample_duration_df['year'], sample_duration_df['duration'])
plt.title('Netflix Movie Duration')
plt.show()
```

```python
netflix_df = pd.read_csv( r"E:\cde\Module_01(Python)\CDE-B3-SH-main\projects\netflix_data.csv")
netflix_df.head(20)
```

```python
netflix_df.isnull().sum()
```

```python
years = netflix_df.release_year.to_list()
durations = netflix_df.duration.to_list()
movie_dict = {
    'years': years,
    'durations': durations
}
```

```python
durations_df = pd.DataFrame(movie_dict)
durations_df.head(10)
```

```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
plt.figure(figsize=(12, 6))
plt.style.use('seaborn-v0_8-whitegrid')
sns.lineplot(x='years' , y='durations' , data=durations_df)
plt.title('Netflex Movies Durations')
plt.xlabel('Release Year')
plt.ylabel('Movies Durations')
plt.show()
```

```python
netflix_movie_df = netflix_df[netflix_df['type'] == 'Movie']
netflix_movies_col_subset = netflix_movie_df[['title', 'country', 'genre', 'release_year', 'duration']]
netflix_movies_col_subset.head()
```

```python
fig = plt.figure(figsize=(12, 6))
plt.scatter(x='years', y='durations', data=durations_df)
plt.title('Netflix Movies Duartions')
plt.show()
```

```python
short_movies = netflix_movie_df[netflix_movie_df['duration'] < 60]
short_movies.head(7)
```

```python

colors = []
for rows, columns in netflix_movies_col_subset.iterrows() :
    if columns['genre'] == 'Children':
        colors.append('red')
    elif columns['genre'] == 'Stand-Up':
        colors.append('green')
    elif columns['genre'] == 'Documentaries':
        colors.append('blue')
    else:
        colors.append('black')
colors[:10]
```

```python
netflix_movie_df.columns
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12, 6))
plt.scatter(x='release_year', y='duration', data=netflix_movie_df, c=colors)
plt.title('Scatter Plot of Years vs 6')
plt.xlabel('Years')
plt.ylabel('Durations')
plt.show()
```
