```python
# Q1: Print alternate elements of a list
lst = list(map(int, input("Enter numbers separated by space: ").split()))
print("Alternate elements:")
for i in range(0, len(lst), 2):
    print(lst[i], end=" ")
```

```python
# Q2: Reverse list without using reverse()
lst = list(map(int, input("Enter numbers separated by space: ").split()))
rev = lst[::-1]   # slicing se reverse
print("Reversed List:", rev)
```

```python
# Q3: Find largest element without using max()
lst = list(map(int, input("Enter numbers separated by space: ").split()))
largest = lst[0]
for num in lst:
    if num > largest:
        largest = num
print("Largest number is:", largest)
```

```python
# Q4: Rotate elements of list
lst = list(map(int, input("Enter numbers separated by space: ").split()))
rotated = lst[1:] + [lst[0]]
print("Rotated List:", rotated)
```

```python
# Q5: Delete a given word from a string
s = input("Enter a string: ")
word = input("Enter word to delete: ")
new_s = s.replace(word, "")
print("String after deletion:", new_s.strip())
```

```python
# Q6: Capitalize each word
s = input("Enter a string: ")
words = s.split()
new_s = " ".join(word.capitalize() for word in words)
print("Output:", new_s)
```

```python
# Q8: Sum of each row
matrix = [
    [2, 11, 7, 12],
    [5, 2, 9, 15],
    [13, 10, 18, 42]
]

for i, row in enumerate(matrix, start=1):
    print(f"Sum of row {i} = {sum(row)}")
```

```python
# Q9: Add two matrices
X = [[1,2,3],
     [4,5,6],
     [7,8,9]]

Y = [[9,8,7],
     [6,5,4],
     [3,2,1]]

result = [[X[i][j] + Y[i][j] for j in range(len(X[0]))] for i in range(len(X))]

print("Matrix after addition:")
for r in result:
    print(r)
```
